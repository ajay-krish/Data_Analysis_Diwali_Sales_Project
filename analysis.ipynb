# Exploratory Data Analysis

## Gender

```python
df.columns
```

```python
ax = sns.countplot(x = 'Gender', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

#### from the above plot we can understand that most of the buyers are female

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=True)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=True)
bx = sns.barplot(x='Gender',y='Amount',data=sales_gen)
for bars in bx.containers:
    bx.bar_label(bars)
```

#### From above graphs we can understand that most of the buyers are female and the purchasing power (total amount spent) of females are greater than males

## Age

```python
df.columns
```

```python
ax = sns.countplot(data=df, x = 'Age Group', hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum()
# Sort the DataFrame by 'Age Group'
sales_age = sales_age.sort_values(by='Age Group')
# Plot the data
sns.barplot(x='Age Group', y='Amount', data=sales_age)
```

#### From the above graphs we can understand that most of the buyers are females of age group 26-35 

## State

```python
df.columns
```

```python
#Total number of orders from top 10 states
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_state, x='State',y='Orders')
```

```python
#Total amount spent from top 10 states
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_state, x='State',y='Amount')
```

#### From the above graphs we can understand that most of the orders are from Uttar pradesh, Maharashtra and Karnataka and even though the amount spent at the states are also in same order till Himachal Pradesh, we can see that Kerala is not in the top amount spent states and Haryana comes in that place. Also Bihar enters the top amount spent states graph. Gujarat retains 10th position in both graphs

## Marital Status

```python
df.columns
```

```python
ax = sns.countplot(data=df, x = 'Marital_Status')
sns.set(rc={'figure.figsize':(5,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_mstatus = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(5,5)})
sns.barplot(data=sales_mstatus, x='Marital_Status',y='Amount', hue='Gender')
```

#### From the above graphs we can understand that most of the buyers are married females and they have high purchasing power

### Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data=df, x = 'Occupation')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occup = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_occup, x='Occupation',y='Amount')
```

#### From the above graphs we can understand that most of the buyers work in IT sector, healthcare and aviation

### Product Category

```python
sns.set(rc={'figure.figsize':(35,7)})
ax = sns.countplot(data=df, x = 'Product_Category')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_prod = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(30,7)})
sns.barplot(data=sales_prod, x='Product_Category',y='Amount')
```

#### From the above graphs we can understand that most sold products are food, clothing &apparels and electronics & gadgets. One thing to note is that even though as per the number of people who ordered the products, clothing products are top followed by food and electronics. But as per the amount spent, Food comes first followed by clothing and electronics products
